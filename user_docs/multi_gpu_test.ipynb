```python
from skimage.io import imread
from pyclesperanto import Clesperanto

cle1 = Clesperanto()
cle1.select_device("RTX")
```

```python
cle2 = Clesperanto()
cle2.select_device("gfx")
```

```python
image = imread("../demo/blobs.tif")

cle1.imshow(image)
```

```python
blurred1 = cle1.gaussian_blur(image, sigma_x=5)

cle1.imshow(blurred1)
```

```python
blurred2 = cle2.gaussian_blur(image, sigma_x=10)

cle2.imshow(blurred2)
```

```python
blurred1 = cle1.gaussian_blur(image, sigma_x=15)

cle1.imshow(blurred1)
```

```python
blurred2 = cle2.gaussian_blur(image, sigma_x=20)

cle2.imshow(blurred2)
```
