```python
from skimage.io import imread
from pyclesperanto import Clesperanto

cle = Clesperanto()
cle.select_device("RTX")
```

```python
image = imread("https://github.com/clEsperanto/clesperanto_example_data/raw/main/Lund_000500_resampled-cropped.tif")

image.shape
```

```python
cle.imshow(image)
```

```python
cle.imshow(cle.maximum_x_projection(image))
```

```python
background_subtracted = cle.difference_of_gaussian(image, sigma2_x=5, sigma2_y=5, sigma2_z=5)

cle.imshow(background_subtracted)
```

```python
binary_image = cle.threshold_otsu(background_subtracted)
label_image = cle.connected_components_labeling_box(binary_image)

cle.imshow(label_image, labels=True)
```

```python
cle.imshow(cle.maximum_x_projection(label_image), labels=True)
```
