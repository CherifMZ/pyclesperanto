```python
from skimage.io import imread
from pyclesperanto import Clesperanto

cle = Clesperanto()
cle.select_device("RTX")
```

```python
image = imread("../demo/blobs.tif")

cle.imshow(image)
```

```python
flipped = image[::-1]
cle.imshow(flipped)
```

```python
summed_image = cle.add_images_weighted(image, flipped)

cle.imshow(summed_image)
```
